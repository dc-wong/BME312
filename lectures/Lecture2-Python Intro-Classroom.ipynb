# Python Toolbox Examples

## NumPy

### Array Creation and Initialization

```python
# Import numpy toolbox
import numpy as np

# Create a NumPy array containing even numbers from 2 to 40
array = np.arange(2,41,2)

# Print result
print(array)

# Print Shape
print(array.shape)
```

```python
# Reshape the array into a 2D array of shape (4, 5)
arr_2d = array.reshape(4, 5)

# Print result
print("Array:")
print(arr_2d)

# Print Shape
print("Shape:")
print(arr_2d.shape)
```

```python
# Reshape 1D array to 2D row array
arr_2dr = array.reshape(1,20)

# Print result
print("Array:")
print(arr_2dr)

# Print Shape
print("Shape:")
print(arr_2dr.shape)
```

```python
# Reshape 1D array to 2D column array
arr_2dc = array.reshape(20,1)

print("Array:")
print(arr_2dc)
print("Shape:")
print(arr_2dc.shape)
```

```python
# Reshape 2D array back to 1D
arr_1d = arr_2dc.reshape(-1)

print("Array:")
print(arr_1d)
print("Shape:")
print(arr_1d.shape)
```

### Indexing and Slicing

```python
# Access and print the element in the second row and third column
element= arr_2d[1, 2]
print("Element in the second row and third column:", element)
```

```python
# Slice the array to extract the sub-array consisting of the first two rows and the last three columns
sub_array = arr_2d[:2, -3:]

print("Sub-array:")
print(sub_array)
```

### Broadcasting

```python
# Create a NumPy array of shape (1, 5) containing five different values
broadcast_array = np.array([[1, 2, 3, 4, 5]])

# Add this 1D array to every row of the 2D array created in exercise 1
result_array = arr_2d + broadcast_array

print("Resulting Array after Broadcasting:")
print(result_array)
```

### Aggregate Function

```python
# Calculate and print the sum, mean, maximum, and minimum values of the entire array
print("Sum:", arr_2d.sum() )
print("Mean:", arr_2d.mean() )
print("Maximum:", arr_2d.max() )
print("Minimum:", arr_2d.min() )

# Print all this in one row
print('The sum of the array is ', np.sum(arr_2d), 'and the mean is ', np.mean(arr_2d))

# Print with no decimal place
print('The sum of the array is %.0f and the mean is %.0f' % (np.sum(arr_2d), np.mean(arr_2d)))
print('The sum of the array is {:.0f}  and the mean is {:.0f}'.format(np.sum(arr_2d), np.mean(arr_2d)))
print(f'The sum of the array is {np.sum(arr_2d):.0f}  and the mean is {np.mean(arr_2d):.0f}')



# Calculate and print the sum along each axis (rows and columns) of the array
print("Sum along rows:", arr_2d.sum(axis=1) )
print("Sum along columns:", arr_2d.sum(axis=0) )
```

### Advanced Operations

```python
# Create 2D broadcast array
broadcast_array2D = np.repeat(broadcast_array,4,axis=0)

#In some versions of python you have to do it this way
broadcast_array2D = np.repeat(broadcast_array.reshape(1,-1),4,axis=0)

print(broadcast_array2D)
```

```python
# Perform element-wise multiplication between the original array and the broadcast array
mul_result = arr_2d * broadcast_array2D
print(mul_result)
```

```python
# Perform element-wise multiplication between the original array and the broadcast array
mul_result = np.multiply(arr_2d, broadcast_array2D)
print(mul_result)
```

```python
# Perform matrix multiplication using the first 4 columns of our 2D array and 1st column of broadcast array
matmul_result = np.matmul(arr_2d[:,:4], broadcast_array2D[:,0])
print(matmul_result)
```

```python
# Perform matrix multiplication using the first 4 columns of our 2D array and 1st column of broadcast array
matmul_result2 = arr_2d[:,:4] @ broadcast_array2D[:,0]
print(matmul_result2)
```

## matplotlib
### Line plot

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Generate a sinusoid with a domain from 0 to 10 with 100 samples data
x = np.arange(0,10.01,.1)
y = np.sin(x)

# Plot a line graph
plt.plot(x, y, label='Sine Wave')

# Label the plot
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Sine wave')

# Show the plot
plt.show()
```

### Scatter Plot

```python
# Generate 100 random data points
x = np.random.randn(100, 1)
y = np.random.randn(100, 1)

rng = np.random.default_rng(312)

x  = rng.normal(size = 100)
y = rng.normal(size = 100)

# Plot a scatter plot with red x markers
plt.scatter(x, y, color = 'red', marker = 'x', label = 'Random Data')

# Label the plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random Scatter Plot')
plt.legend()
```

### Histogram

```python
# Generate random data
data = np.random.randn(1000)

# Plot a histogram
plt.hist(data, bins = 30, color = 'skyblue', edgecolor = 'black', alpha = .7)

# Label the plot
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

# Show the plot
plt.show()
```

### Bar plot

```python
# Define categories and values
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 32] 

# Plot a bar plot
plt.bar(categories, values, color = 'green')

# Label the plot
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Plot')

# Limit the plot to 0 t0 60 on the y-axis
```

### Error Bar

```python
error_values = [3, 5, 6, 2]

# Plot an error bar plot
plt.errorbar(categories, values, error_values, color='black')

# Customize the plot
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Error Bar Plot')
plt.ylim(0, 60)
```

```python
# Plot a bar plot & Error bar plot
plt.bar(categories, values, color='green')
plt.errorbar(categories, values,error_values, color='black', fmt='none', capsize=10)

# Customize the plot
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Plot')
plt.ylim(0, 70)
```

## Seaborn

```python
#import seaborn
import seaborn as sns

# Load a sample dataset (e.g., Iris dataset)
iris_df = sns.load_dataset('iris')
```

### Scatter plot

```python
# Create a scatter plot using Seaborn's scatterplot function
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species')

# Customize the plot
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')

# Show the plot
plt.show()
```

### Histogram

```python
# Create a histogram using Seaborn's histplot function
sns.histplot(data=iris_df, x='petal_length', kde=True, color='skyblue')

# Customize the plot
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')

# Show the plot
plt.show()
```

### Boxplot

```python
# Create a box plot using Seaborn's boxplot function
sns.boxplot(data = iris_df, x='species', y = 'petal_length', palette='Set2', hue='species')

# Customize the plot
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Width across Species')

# Show the plot
plt.show()
```

### Heatmap

```python
# Compute the correlation matrix
correlation_matrix = iris_df.corr(numeric_only = True)

# Create a heatmap using Seaborn's heatmap function
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Customize the plot
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```

The enumerate function gives you the index as you loop through a for loop
